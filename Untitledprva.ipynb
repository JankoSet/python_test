```python
# % tam pomeni, da bomo gledali ostanek pri deljenju
# == je enako
# != ni enako
n=input("Choose a number higher than 1.")
n=int(n)

odd=0
even=0
while n!=1 :
    if n%2==0:
        n=n//2
    else:
        n=3*n+1
    print(n)
    
    if n%2==0:
        odd=odd+1
    else:
        even=even+1
    print(f"Total number of odd 'n' is {odd}")
    print(f"Total number of even 'n' is {even}")
    print(f"Total length of the sequence is {odd+even}")
```

```python
# i je tradicionalen števec
from math import *
from random import *

N=1000000

i=0
within=0
while i<N:
    x=uniform(-1, 1)
    y=uniform(-1, 1)
    if sqrt(x**2+y**2)<1:
        within=within+1
    i=i+1

p=within/N
print(f"Estimate for pi is {4*p}")
```
