```python
#preštejmo adenine ("a") v sekvencah v datotei qwerty-dna.txt:
sekvence1=open("qwerty-dna.txt")

for line in sekvence1:
    gene, sequence=line.split()
    
    cnt=0
    for char in sequence:
        if char == "a":
            cnt=cnt+1
    print(gene, sequence, cnt)
```

```python
#Sedaj pa izpišimo le gen z največ adeninov in število adeninov v njegovi sekvenci.
#Od prej:
#biggest=0
#for line in open ("teze.txt"):
#    if int(line)>biggest:
#        biggest=int(line)
        
#print(biggest)
sekvence1=open("qwerty-dna.txt")
highest_as = -1

for line in sekvence1:
    name, sequence=line.split()
    a = sequence.count("a")
    if a > highest_as:
        highest_as = a
        gene = name

print(gene, highest_as)
```

```python
#Najvišja frekvenca:

seq

counts = {}
for i in range(0, len(seq)-3):
    triplet = (seq[i:i+3])
    if triplet not in counts:
        counts[triplet] = 0
    counts[triplet] = counts[triplet] + 1

high_freq = -1
high_freq_triplet = ""
    
for triplet in counts:
    freq = counts[triplet]
    if freq > high_freq:
        high_freq = freq
        high_freq_triplet = triplet
print(high_freq_triplet, high_freq)
```
